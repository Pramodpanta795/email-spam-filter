```python
#frequency count
text = "programming"

freq = {}
for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)

#bar Chart
from collections import Counter
import matplotlib.pyplot as plt

# categorical data
data = ['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C', 'B', 'A']

# empirical distribution
counts = Counter(data)
total = len(data)
categories = list(counts.keys())
frequencies = list(counts.values())
probabilities = [f / total for f in frequencies]

# BAR CHART
plt.figure()
plt.bar(categories, frequencies)
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.title("Bar Chart of Categorical Data")
plt.show()

# PARETO CHART
sorted_items = counts.most_common()
sorted_categories = [item[0] for item in sorted_items]
sorted_frequencies = [item[1] for item in sorted_items]
cumulative_percent = []
cum_sum = 0

for f in sorted_frequencies:
    cum_sum += f
    cumulative_percent.append(cum_sum / total * 100)

plt.figure()
plt.bar(sorted_categories, sorted_frequencies)
plt.plot(sorted_categories, cumulative_percent, marker='o')
plt.xlabel("Category")
plt.ylabel("Frequency / Cumulative Percentage")
plt.title("Pareto Chart")
plt.show()

from collections import Counter
import matplotlib.pyplot as plt

# categorical data
data = ['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C', 'B', 'A']

# empirical distribution
counts = Counter(data)
labels = list(counts.keys())
sizes = list(counts.values())

# PIE CHART
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart of Categorical Data")
plt.axis('equal')
plt.show()

```
